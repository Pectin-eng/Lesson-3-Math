```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Y = np.arange(-50, 50, 1)
X, Y = np.meshgrid(X, Y)
Z2 = 1 - 2*X - 3*Y
Z1 = (-4 - (2/3)*X - Y) * 3
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```
