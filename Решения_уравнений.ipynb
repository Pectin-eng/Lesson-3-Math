```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

x = np.linspace(-5, 5, 200)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 + 1)
plt.show()

def equation(p):
  x, y = p
  return (y - x**2 + 1, np.exp(x) + x - x * y - 1)

x, y = fsolve(equation, (44, 6))
print(x, y)

x, y = fsolve(equation, (3, 7))
print(x, y)

x, y = fsolve(equation, (-2, 3))
print(x, y)
```
